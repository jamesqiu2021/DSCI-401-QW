1. Using the data whately 2015 from the macleish package, create an interactive plot using plotly (or ggplotly) displaying time (in days) on the x-axis
and temperature on the y-axis with three lines: one for the high temperature of the day, one for the average temperature of the day, and one for
the low temperature of the day. A csv version of the file can be found here:
https://www.dropbox.com/s/m2nt50qanpijp0m/whately2015.csv?dl=0

```python
import pandas as pd
import numpy as np
```

```python
wht2015=pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/whately_2015.csv")
wht2015.head(2)
```

```python
# data wrangling, date covert.
from pickle import FALSE
from numpy.core.arrayprint import format_float_scientific
from pandas._libs.lib import fast_unique_multiple_list_gen
wt2=wht2015
wt2['when'] = pd.to_datetime(wt2['when'])
wt2['date'] = wt2['when'].dt.strftime('%Y-%m-%d')
wt3=wt2.groupby('date',as_index=False).agg({'temperature':['max','mean','min']})
wt3.head(5)
```

```python
# set the columns at the same level
wt4=wt3
wt4.columns = wt4.columns.get_level_values(0)
wt4.head()
```

```python
# rename the columns
wt6=wt3
wt6.columns=['date','high_temp','ave_temp',"low_temp"]
wt6.head()
```

```python
#pivot longer
wt7=pd.melt(wt6,id_vars=["date"],value_vars=["high_temp","ave_temp","low_temp"],var_name="temp_type",value_name="temperature")
wt7.head(5)
```

```python
#plotly
import plotly
import plotly.express as px
fig=px.line(wt7,x='date',y="temperature",color="temp_type")
fig.show()
```
