```python
import pandas as pd
airlines = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/airlines.csv")
flights = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/flights.csv")
airports = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/airports.csv")
batting = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/Batting.csv")
people = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/People.csv")
Pitching = pd.read_csv("https://raw.githubusercontent.com/jamesqiu2021/DSCI-401-QW/main/Pitching.csv")
planes = pd.read_csv("https://raw.githubusercontent.com/tidyverse/nycflights13/main/data-raw/planes.csv")
```

#1(a) (10 points) Name every player in baseball history who has accumulated at least 300 home runs (HR) and at least 300 stolen bases (SB).You can find the first and last name of the player in the People data frame. Join this to your result along with the total home runs and total bases stolen for each of these elite players.

```python
from numpy.lib.shape_base import column_stack
column_stack(people)
```

```python
from pandas.core.resample import PeriodIndexResampler
from pandas.core.reshape.merge import merge
q = batting.loc[:,["playerID","HR","SB"]].groupby("playerID").sum()
q1=q[q.HR>300][q.SB>300]
people["name"]=people["nameFirst"]+people["nameLast"]
people_1=people[["playerID","name"]]
merge_q = pd.merge(people_1,q1,on="playerID",how="inner")
merge_q


```

#1(b) (10 points) Similarly, name every pitcher in baseball history who has accumulated at least 300 wins (W) and at least 3,000 strikeouts (SO).

```python
q2 = Pitching.loc[:,["playerID","W","SO"]].groupby("playerID").sum()
q3=q2[q2.W>300][q2.SO>3000]
people["name"]=people["nameFirst"]+people["nameLast"]
people_1=people[["playerID","name"]]
merge_q2 = pd.merge(people_1,q3,on="playerID",how="inner")
merge_q2
```

#1(c) (10 points) Identify the name and year of every player who has hit at least 50 home runs in a single season. Which player had the lowest batting average in that season?

```python
q1c1=batting[["yearID","playerID","HR"]][batting.HR>=50]
merge_q3 = pd.merge(q1c1,people_1,on="playerID",how="inner")
merge_q3
```

```python
merge_q4 = pd.merge(merge_q3,batting,on=("playerID","yearID"),how="inner")
merge_q4["batting_average"]=merge_q4["H"]/merge_q4["AB"]
merge_q4[["yearID","playerID","name","batting_average"]].sort_values(by="batting_average",ascending=True).head(5)
```

#2(a) (10 points) What is the oldest plane (specified by the tailnum variable) that flew from New York City airports in 2013?
